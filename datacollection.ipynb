# data from API


```python
import requests
import csv
import bs4
import pandas as pd

resp = requests.get("https://gorest.co.in/public/v2/posts")
print(resp)
emps = resp.json()
emps
```

```python
with open("data.csv",'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'user_id', 'title', 'body'])
    writer.writeheader()
    for emp in emps:
        d_id = emp['id']
        uid = emp['user_id']
        title = emp['title']
        body = emp['body']
        writer.writerow(emp)
df = pd.read_csv('data.csv')
df
```
