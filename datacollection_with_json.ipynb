```python
import requests
import csv
import pandas as pd

resp = requests.get("https://gorest.co.in/public/v2/posts")
resp_comments = requests.get("https://gorest.co.in/public/v2/comments")
resp_todos = requests.get("https://gorest.co.in/public/v2/todos")
resp_users = requests.get("https://gorest.co.in/public/v2/users")
emps = resp.json()
comments = resp_comments.json()
todos = resp_todos.json()
users = resp_users.json()
```

```python
with open("data.csv",'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'user_id', 'title', 'body'])
    writer.writeheader()
    for emp in emps:
        d_id = emp['id']
        uid = emp['user_id']
        title = emp['title']
        body = emp['body']
        writer.writerow(emp)
df = pd.read_csv('data.csv')
df
```

```python
# users
with open("data_users.csv",'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'name', 'email', 'gender','status'])
    writer.writeheader()
    for user in users:
        _id = user['id']
        name = user['name']
        email = user['email']
        gender = user['gender']
        status = user['status']
        writer.writerow(user)
df = pd.read_csv('data_users.csv')
df
```

```python
# Comments
with open("data_comments.csv",'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'post_id', 'name', 'email', 'body'])
    writer.writeheader()
    for comment in comments:
        _id = comment['id']
        post_id = comment['post_id']
        name = comment['name']
        email = comment['email']
        body = comment['body']
        writer.writerow(comment)
df = pd.read_csv('data_comments.csv')
df
```

```python
# todos
with open("data_todos.csv",'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=['id', 'user_id', 'title', 'due_on', 'status'])
    writer.writeheader()
    for todo in todos:
        _id = todo['id']
        user_id = todo['user_id']
        title = todo['title']
        due_on = todo['due_on']
        status = todo['status']
        writer.writerow(todo)
df = pd.read_csv('data_comments.csv')
df
```
